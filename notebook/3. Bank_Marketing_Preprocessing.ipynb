### Removing the 'duration' Column – Reasoning

The **'duration'** column represents the duration of the **last call** and is only known after the call has ended. At that point, the **subscription result (y)** (whether the customer subscribed or not) is already determined. Furthermore, when **duration = 0**, the result is always **y = 'no'**.

This creates a risk of **data leakage**, as the feature **'duration'** could potentially reveal information about the target variable **'y'** (subscription result). Removing this column helps ensure the integrity of the model by preventing the model from unintentionally using future information to predict outcomes.

**Insight:**  
By removing the **'duration'** column, we eliminate the risk of data leakage and ensure that the model is based solely on information available before the outcome is known.

```python
df.drop(columns='duration', inplace=True)
```

## Create a pipeline for handling text-related errors

- **JobAdminCleaner** (`clean_job_admin`): Fix the input error where "admin." is mistakenly entered as "admin" in the **'job'** column.
- **EducationPunctuationCleaner** (`clean_education_punct`): Replace the period "." with a space " " in the **'education'** column (e.g., "basic.9y" → "basic 9y", "unknown." → "unknown "). This helps standardize values and correct input errors.
- Convert text to numeric values in the **'month'** and **'days_of_week'** columns.

```python
class PdaysHandler(BaseEstimator, TransformerMixin):
    """
    Handle the value 999 in the 'pdays' column.
    1. Create a new binary column 'was_contacted_previously'.
    2. Replace the value 999 in the original 'pdays' column with NaN (or other values).
    """
    def __init__(self, pdays_col='pdays', new_col_suffix='_contacted_status', replace_999_with=np.nan):
        self.pdays_col = pdays_col
        self.new_col_name = f"{pdays_col}{new_col_suffix}"
        self.replace_999_with = replace_999_with

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X_transformed = X.copy()
        if self.pdays_col in X_transformed.columns:
            print(f"Processing column '{self.pdays_col}': Creating column '{self.new_col_name}' and replacing 999 with {self.replace_999_with}")

            # Step 1: Create a new column (before modifying the original column)
            X_transformed[self.new_col_name] = np.where(X_transformed[self.pdays_col] == 999, 0, 1)

            # Step 2: Explicitly convert the original pdays column to float before replacing
            # This addresses FutureWarning
            if pd.api.types.is_numeric_dtype(X_transformed[self.pdays_col]):  # Only convert if numeric
                X_transformed[self.pdays_col] = X_transformed[self.pdays_col].astype(float)
            else:
                # If it's not numeric, try to convert it to numeric, raise error if not possible
                try:
                    X_transformed[self.pdays_col] = pd.to_numeric(X_transformed[self.pdays_col], errors='raise').astype(float)
                except (ValueError, TypeError):
                    warnings.warn(f"Warning: Unable to convert column '{self.pdays_col}' to float. Skipping replacement of 999.")
                    return X_transformed  # Return without changing pdays if conversion fails

            # Step 3: Replace 999 with the desired value (now the type is compatible)
            X_transformed.loc[:, self.pdays_col] = X_transformed[self.pdays_col].replace(999.0, self.replace_999_with)  # Replace 999.0 (since it's now float)

        else:
            warnings.warn(f"Warning: Column '{self.pdays_col}' not found. Skipping the PdaysHandler step.")
        return X_transformed

# --- Run Step 5 again with the updated Transformer ---
# (Reuse df from Step 4)
print("\n--- Step 4 (Repeat): Handling 'pdays' column with the updated Transformer ---")
print("Values in the 'pdays' column before processing:\n", df[['pdays']].head())  # df here still has 'pdays' as int from the initial creation
pdays_handler_fixed = PdaysHandler(pdays_col='pdays', replace_999_with=np.nan)
df_cleaned = pdays_handler_fixed.fit_transform(df_cleaned)  # Create new df to see the difference
print("\nValues in the relevant columns after processing (updated):\n", df_cleaned[['pdays', 'pdays_contacted_status']].head())
print("Unique values in the 'pdays' column after processing (updated):", df_cleaned['pdays'].unique())
print("Unique values in the new 'pdays_contacted_status' column (updated):", df_cleaned['pdays_contacted_status'].unique())
print("Data type of 'pdays' column after processing (updated):", df_cleaned['pdays'].dtype)
```
